### Tarefa 1:
• Crie um jupyter notebook com os imports necessários e um código para gerar os
gráficos referentes a base de dados de março, abril, maio, junho e dezembro
(numa célula só ou em várias).

```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando a base de dados
df = pd.read_csv('SINASC_RO_2019.csv')
```

```python
# Convertendo a coluna de data de nascimento para o tipo datetime e extraindo o mês
df['DTNASC'] = pd.to_datetime(df['DTNASC'])
df['DTNASC'] = df['DTNASC'].dt.month
```

```python
# Filtrando os dados para os meses desejados
meses = [3, 4, 5, 6, 12] # os meses agora são números, não strings
df = df[df['DTNASC'].isin(meses)]

```

```python
# Gerando os gráficos para cada mês
for mes in meses:
    df_mes = df[df['DTNASC'] == mes]
    plt.figure(figsize=(10,6))
    df_mes['APGAR5'].value_counts().plot(kind='bar')
    plt.title(f'Gráfico para o mês {mes}')
    plt.show()
```

### Tarefa 2:
• Como que você poderia fazer para deixar isso mais automatizado? Descreva na
última célula do jupyter notebook o que ajudaria na hora de gerar novos gráficos
pros meses/anos seguintes.

```python
# Função para automatizar o processo
def gerar_graficos(arquivo, meses):
    df = pd.read_csv(arquivo)
    df['DTNASC'] = pd.to_datetime(df['DTNASC'])
    df['DTNASC'] = df['DTNASC'].dt.month
    df = df[df['DTNASC'].isin(meses)]
    for mes in meses:
        df_mes = df[df['DTNASC'] == mes]
        plt.figure(figsize=(10,6))
        df_mes['APGAR5'].value_counts().plot(kind='bar')
        plt.title(f'Gráfico para o mês {mes}')
        plt.show()

# Agora podemos chamar essa função com diferentes parâmetros
gerar_graficos('SINASC_RO_2019.csv', [3, 4, 5, 6, 12])
```
