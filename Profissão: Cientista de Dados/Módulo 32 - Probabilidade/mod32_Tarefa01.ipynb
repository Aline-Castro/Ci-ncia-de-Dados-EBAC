# Probabilidade - Tarefa

```python
# IMPORTANDO AS BIBLIOTECAS NECESSRIAS
import numpy as np
from scipy import stats
from scipy.stats import norm
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import scipy.stats as stats


%matplotlib inline
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parmetros da distribuio
media = 15 # mdia em Kg
desvio_padrao = 0.1 # desvio padro em kg
```

```python
# Calculando a probabilidade de um pacote ter mais que 15.2Kg
prob = 1 - stats.norm.cdf(15.2, loc = media, scale = desvio_padrao)

print(f'A probabilidade de um pacote ter mais que 15,2Kg : {prob*100:.2f}%')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Calculando o percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o peso dos pacotes  {percentil_95:.4f} kg')

```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# Valores em Y
Y=np.linspace(-6, 6, 200)
```

```python
# Calculando as probabilidades para Y1 e Y2
prob_Y1 = (stats.norm(loc=0, scale=1)).cdf(Y) # distribuio normal padro
prob_Y2 = (stats.t(df=5)).cdf(Y) # distribuio t-student com 5 graus de liberdade

# Imprima os resultados
for i in range(len(Y)):
    print(f'Para Y1, a probabilidade de ser menor que {Y[i]:.2f}  {prob_Y1[i]:.4f}')
    print(f'Para Y2, a probabilidade de ser menor que {Y[i]:.2f}  {prob_Y2[i]:.4f}')
    print('-' * 40)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Grfico de linhas para representar as probabilidades acumulativas
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_Y1, label='Y1 - Normal Padro')
plt.plot(Y, prob_Y2, label='Y2 - t-Student (df=5)')

# Adicionando rtulos e legendas
plt.title('Probabilidade Acumulativa para Y1 e Y2', color = 'navy')
plt.xlabel('Valores de Y', color = 'navy')
plt.ylabel('Probabilidade Acumulativa', color = 'navy')
plt.legend()
plt.grid(True)
plt.show()
```

5. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Valores em Y
Y = np.linspace(0, 1, 20)
```

```python
# Calcule os percentis para Y1 e Y2
percentis_Y1 = (stats.norm(loc=0, scale=1)).ppf(Y) # Distribuio Normal Padro
percentis_Y2 = (stats.t(df=5)).ppf(Y) # Distribuio t-Student com 5 graus de liberdade

# Imprima os resultados
for i in range(len(Y)):
    print(f'Para Y1, o percentil para {Y[i]:.2f}  aproximadamente: {percentis_Y1[i]:.4f}')
    print(f'Para Y2, o percentil para {Y[i]:.2f}  aproximadamente: {percentis_Y2[i]:.4f}')
    print('-' * 40)
```

6. Faa um grfico com os dados do item anterior.

```python
# Grfico para representar os percentis
plt.figure(figsize=(10, 6))
plt.plot(Y, percentis_Y1, label='Y1 - Normal Padro', color='blue')
plt.plot(Y, percentis_Y2, label='Y2 - t-Student (df=5)', color='orange')

# Adicionando rtulos e legendas
plt.title('Percentis para Y1 e Y2', color = 'navy')
plt.xlabel('Probabilidades', color = 'navy')
plt.ylabel('Valores de Y', color = 'navy')
plt.legend()
plt.grid(True)
plt.show()
```
