```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('sinasc_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Função para criar gráfico de barras
def plot_bar_chart(data, title, x_label, y_label, color):
    plt.figure(figsize=(6, 4))
    data.value_counts().sort_index().plot(kind='bar', color=color)
    plt.title(title, color='blue')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xticks(rotation=45)
    plt.show()

# Gráfico de barras para APGAR1
plot_bar_chart(df['APGAR1'], 'Distribuição de APGAR1', 'APGAR1', 'Contagem', 'skyblue')

# Gráfico de barras para APGAR5
plot_bar_chart(df['APGAR5'], 'Distribuição de APGAR5', 'APGAR5', 'Contagem', 'lightcoral')

# Gráfico de barras para SEXO
plot_bar_chart(df['SEXO'], 'Distribuição de SEXO', 'SEXO', 'Contagem', 'plum')

# Gráfico de barras para GESTACAO
plot_bar_chart(df['GESTACAO'], 'Distribuição de GESTACAO', 'GESTACAO', 'Contagem', 'lightgreen')

# Gráfico de barras para ESTCIVMAE
plot_bar_chart(df['ESTCIVMAE'], 'Distribuição de ESTCIVMAE', 'ESTCIVMAE', 'Contagem', 'lightblue')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Média de APGAR5 para cada valor único de CONSULTAS
df_apgar5 = df.groupby('CONSULTAS')['APGAR5'].mean()

# Cria o gráfico de barras
plt.figure(figsize=(6,3))
plt.bar(df_apgar5.index, df_apgar5.values)

# Adiciona título e rótulos aos eixos
plt.title('APGAR5 médio por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('APGAR5 médio')

# Mostra o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Lista das colunas
columns = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

# Figura com 2 linhas e 2 colunas de subplots
fig, axes = plt.subplots(2, 2, figsize=(10,10))

# Espaçamento entre os subplots
plt.subplots_adjust(wspace=0.5, hspace=1)

# Gráfico para cada coluna
for i, column in enumerate(columns):
    # Frequência de cada valor único na coluna
    value_counts = df[column].value_counts()

    # Subplot atual
    ax = axes[i//2, i%2]

    # Gráfico de barras no subplot atual
    ax.bar(value_counts.index, value_counts.values)

    # Título e rótulos aos eixos
    ax.set_title(f'Frequência de {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequência')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Mostra a figura
plt.show()
```
