````python
# Importando as bibliotexas necessárias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

````python
# Carregando o arquivo CSV em um DataFrame
df = pd.read_csv('previsao_de_renda.csv')
df.head() # mostra as 5 primeiras linhas do dataframe
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python

# Certificando que a coluna data seja do tipo datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

# Considerando 'sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia' como variáveis qualitativas
qualitative_vars = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

fig, axs = plt.subplots(len(qualitative_vars), 1, figsize=(15, 30))

for i, var in enumerate(qualitative_vars):
    ax = axs[i]

    # Agrupando os dados por data e categoria
    grouped = df.groupby([df['data_ref'].dt.to_period('M'), var]).size()

    # Desempacotando os índices de multi-nível
    grouped = grouped.reset_index()
    grouped.columns = ['mes_ano', var, 'contagem']

    # Pivot dos dados para que cada categoria seja uma coluna
    pivot_df = grouped.pivot(index='mes_ano', columns=var, values='contagem')

    # Substituindo os valores NaN por 0
    pivot_df = pivot_df.fillna(0)

    # Criando o gráfico de barras
    pivot_df.plot(kind='bar', ax=ax)

    ax.set_title(var, color='blue')
    ax.set_xlabel('Mês e Ano', color='purple')
    ax.set_ylabel('Contagem', color='purple')
    ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# Resultado
plt.tight_layout()
plt.show()
````



> ##### **Análise**:
No conjunto de dados analisado, a variável qualitativa "escolaridade" apresentou um comportamento notável. A partir de setembro de 2015, nota-se um aumento significativo na frequência de registros que indicam a conclusão do ensino superior, ao mesmo tempo em que observamos uma redução correspondente na frequência de registros com escolaridade secundária. Após essa mudança, a distribuição da variável "escolaridade" permaneceu relativamente estável até o final do período de análise. Esta transformação significativa na distribuição da variável "escolaridade" ao longo do tempo é um aspecto que merece uma investigação mais detalhada, pois pode ter implicações de grande relevância para o desenvolvimento de modelos de análise de dados.



2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
plt.rc('figure', figsize=(16,4))

for col in qualitative_vars:
    fig, axes = plt.subplots(1,2)
    fig.subplots_adjust(wspace=.6)

    tick_labels = df['data_ref'].map(lambda x: x.strftime('%b/%Y')).unique()

    # barras empilhadas:
    df_crosstab = pd.crosstab(index=df['data_ref'], columns=df[col],normalize='index')
    ax0 = df_crosstab.plot.bar(stacked=True, ax=axes[0])
    ax0.set_xticklabels(labels=tick_labels, rotation=45)
    legend0 = axes[0].legend(bbox_to_anchor=(1,.5), loc=6, title=f"'{col}'")
    plt.setp(legend0.get_title(), color='purple')
    axes[0].set_title(f"'{col}'", color='blue')
    axes[0].set_xlabel('Mês e Ano', color='purple')
    axes[0].set_ylabel('Proporção', color='purple')

    # perfis médios no tempo:
    ax1 = sns.pointplot(x='data_ref', y='renda', hue=col, data=df, dodge=True, ci=95, ax=axes[1])
    ax1.set_xticklabels(labels=tick_labels, rotation=45)
    legend1 = axes[1].legend(bbox_to_anchor=(1,.5), loc=6, title=f"'{col}'")
    plt.setp(legend1.get_title(), color='purple')
    axes[1].set_title(f"'{col}'", color='blue')
    axes[1].set_xlabel('Mês e Ano', color='purple')
    axes[1].set_ylabel('Renda Média', color='purple')


# Resultado
    plt.show()
````

> ##### **Análise**:
Identifica-se uma alteração substancial na variável "escolaridade" durante o período de análise. Essa mudança resultou em uma diferença notável que abrangeu quase a metade do período sob investigação. No entanto, é importante ressaltar que, em contraste com a variável "escolaridade", todas as outras variáveis qualitativas mantiveram uma distribuição constante ao longo do tempo. Dessa forma, apode-se afirmar que essas variáveis podem ser integradas eficazmente em um modelo de análise de dados, devido à sua estabilidade temporal.
