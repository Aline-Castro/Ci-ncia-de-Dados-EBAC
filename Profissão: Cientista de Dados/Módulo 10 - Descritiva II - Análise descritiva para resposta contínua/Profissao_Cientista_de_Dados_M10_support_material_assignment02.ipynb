````python
# Importando as bibliotexas necessárias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

%matplotlib inline
````

````python
# Carregando o arquivo CSV em um DataFrame
df = pd.read_csv('previsao_de_renda.csv')
df.head() # mostra as 5 primeiras linhas do dataframe
````

````python
# Configurando o tamanho padrão das figuras
plt.rc('figure', figsize=(8,8))

# Cria uma grade de subplots 2x2
fig, axes = plt.subplots(2,2)

# Cria um gráfico de pontos para 'posse_de_imovel' vs 'renda'
ax1 = sns.pointplot(x='posse_de_imovel', y='renda', data=df, dodge=True, ax=axes[0,0], color='blue')
ax1.set_xlabel('Posse de Imóvel', color='blue')
ax1.set_ylabel('Renda', color='blue')

# Cria um gráfico de pontos para 'posse_de_veiculo' vs 'renda'
ax2 = sns.pointplot(x='posse_de_veiculo', y='renda', data=df, dodge=True, ax=axes[0,1], color='green')
ax2.set_xlabel('Posse de Veículo', color='blue')
ax2.set_ylabel('Renda', color='blue')

# Cria um gráfico de pontos para 'posse_de_imovel' vs 'renda', colorido por 'posse_de_veiculo'
ax3 = sns.pointplot(x='posse_de_imovel', y='renda',  hue='posse_de_veiculo', data=df, dodge=True, ax=axes[1,0], palette='Set1')
ax3.set_xlabel('Posse de Imóvel', color='blue')
ax3.set_ylabel('Renda', color='blue')

# Cria um gráfico de pontos para 'posse_de_veiculo' vs 'renda', colorido por 'posse_de_imovel'
ax4 = sns.pointplot(x='posse_de_veiculo', y='renda', hue='posse_de_imovel', data=df, dodge=True, ax=axes[1,1], palette='Set1')
ax4.set_xlabel('Posse de Veículo', color='blue')
ax4.set_ylabel('Renda', color='blue')

# Colocando a legenda fora dos gráficos
for ax in [ax1, ax2, ax3, ax4]:  ax.legend(bbox_to_anchor=(1.05, 1), loc=2)

# Exibe os gráficos
plt.tight_layout()
plt.show()
````


> Com base na análise dos gráficos, pode-se concluir que ambas as variáveis qualitativas, 'posse_de_imovel' e 'posse_de_veiculo', apresentam uma discriminação estatisticamente significativa da renda. No entanto, a variável 'posse_de_veiculo' parece ter uma maior relevância para a predição de renda. Isso é evidenciado pelo maior distanciamento do intervalo de confiança entre os indivíduos que possuem um veículo e aqueles que não possuem.

> Ao analisar os gráficos na segunda linha, que fornecem uma visão mais detalhada, notamos que a variável 'posse_de_imovel' tem uma menor significância. Isso ocorre porque, para os indivíduos que possuem um veículo, a renda parece ser semelhante, independentemente da posse de imóvel.

> Essas conclusões destacam a importância de considerar várias variáveis ao tentar prever a renda. Embora ambas as variáveis 'posse_de_imovel' e 'posse_de_veiculo' sejam estatisticamente significativas, a relevância de cada uma pode variar. Portanto, é crucial levar em conta essas diferenças ao desenvolver modelos preditivos ou conduzir análises mais profundas.
