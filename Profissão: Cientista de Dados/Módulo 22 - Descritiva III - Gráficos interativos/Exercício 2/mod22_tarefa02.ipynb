````python
# Importando as bibliotecas necessárias
import pandas as pd
import plotly.graph_objects as go
````

## 1. Carregue os dados da tarefa anterior

````python
# Carregando os dados do arquivo CSV
df = pd.read_csv('ativos.csv')

````

Obs.: Os três ativos escolhidos foram:<br>

AAPL: ativo da Apple Inc., uma empresa multinacional americana de tecnologia sediada em Cupertino, Califórnia1. <br>
MSFT: ativo da Microsoft Corporation, uma corporação multinacional de tecnologia americana sediada em Redmond, Washington2.<br>
GOOGL: ativo da Google LLC, uma empresa multinacional de tecnologia americana.

## 2. Média móvel

### 2.1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu


````python
# Convertendo a coluna 'Data' para o tipo datetime
df['Data'] = pd.to_datetime(df['Data'])


# Calculando a média móvel em uma janela de 90 dias para cada ativo
for ativo in df['Ativo'].unique():
    df.loc[df['Ativo'] == ativo, 'MediaMovel'] = df[df['Ativo'] == ativo]['Adj Close'].rolling(window=90).mean()
````

### 2.2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima


````python
# Calculando o desvio padrão em uma janela de 90 dias para cada ativo
for ativo in df['Ativo'].unique():
    df.loc[df['Ativo'] == ativo, 'DesvioPadrao'] = df[df['Ativo'] == ativo]['Adj Close'].rolling(window=90).std()
````

### 2.3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base


````python
# Criando o gráfico com Plotly
fig1 = go.Figure()
for ativo in df['Ativo'].unique():
    fig1.add_trace(go.Scatter(x=df[df['Ativo'] == ativo]['Data'],
                              y=df[df['Ativo'] == ativo]['MediaMovel'],
                              mode='lines',
                              name=ativo))

# Ajustando a formatação do eixo x para mostrar todos os meses
fig1.update_xaxes(dtick="M1", tickformat="%b\n%Y")

fig1.show()
````

### 2.4. Monte um gráfico de linhas para os desvios padrão em janela móvel


````python
# Criando o gráfico com Plotly
fig2 = go.Figure()
for ativo in df['Ativo'].unique():
    fig2.add_trace(go.Scatter(x=df[df['Ativo'] == ativo]['Data'],
                              y=df[df['Ativo'] == ativo]['DesvioPadrao'],
                              mode='lines',
                              name=ativo))

# Ajustando a formatação do eixo x para mostrar todos os meses
fig2.update_xaxes(dtick="M1", tickformat="%b\n%Y")

fig2.show()
````

### 2.5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Identificando o período de maior volatilidade para cada ativo
for ativo in df['Ativo'].unique():
    periodo_maior_volatilidade_media_movel = df[df['Ativo'] == ativo]['Data'][df[df['Ativo'] == ativo]['MediaMovel'].idxmax()]
    periodo_maior_volatilidade_desvio_padrao = df[df['Ativo'] == ativo]['Data'][df[df['Ativo'] == ativo]['DesvioPadrao'].idxmax()]

    print(f'Para o ativo {ativo}, o período de maior volatilidade de acordo com a média móvel é: {periodo_maior_volatilidade_media_movel}')
    print(f'Para o ativo {ativo}, o período de maior volatilidade de acordo com o desvio padrão é: {periodo_maior_volatilidade_desvio_padrao}')
````

**ANÁLISE:**<br>
Com base nos resultados obtidos e analisando os gráficos pode-se aferir:

Os ativos da Apple (AAPL), Microsoft (MSFT) e Google (GOOGL) apresentaram períodos de maior volatilidade em diferentes momentos ao longo dos últimos dois anos.

>Para a **Apple**, o período de maior volatilidade, de acordo com a média móvel, foi em 25 de setembro de 2023, enquanto o período de maior volatilidade, de acordo com o desvio padrão, foi em 6 de julho de 2022. Isso sugere que o preço das ações da Apple teve grandes variações nesses períodos.

>Para a **Microsoft**, ambos os períodos de maior volatilidade ocorreram em 2023, com a média móvel atingindo seu pico em 25 de setembro e o desvio padrão atingindo seu pico em 29 de junho. Isso indica que o preço das ações da Microsoft foi particularmente volátil nesses momentos.

>Para o **Google**, o período de maior volatilidade, de acordo com a média móvel, foi muito mais cedo, em 4 de março de 2022. No entanto, o desvio padrão atingiu seu pico mais tarde, em 23 de junho de 2023. Isso sugere que houve uma grande variação no preço das ações do Google nesses períodos.

Esses resultados destacam a importância de monitorar a volatilidade ao investir em ações. Períodos de alta volatilidade podem apresentar oportunidades para os investidores, mas também podem representar um risco maior. Portanto, é crucial para os investidores entenderem e monitorarem a volatilidade dos ativos em que estão investindo.

> Também é interessante notar que no ***período 2023-09-25***, tanto Apple quanto Microsoft tiveram maior **volatilidade** de acordo com a média móvel.

> E de acordo com o **desvio padrão**, Microsoft e Google tiveram a maior volatilidade em períodos próximos, sendo MSFT no dia 29/06/2023 e GOOGL no dia 23/06/2023
