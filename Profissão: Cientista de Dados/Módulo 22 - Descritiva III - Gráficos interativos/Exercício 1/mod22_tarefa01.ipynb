```python
# Importando as bibliotecas necessárias
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime, timedelta
```

## 1. Baixando dados de ativos


### 1.1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.




> Os três ativos escolhidos foram: <br>
* **AAPL**: ativo da Apple Inc., uma empresa multinacional americana de tecnologia sediada em Cupertino, Califórnia1.
* **MSFT**: ativo da Microsoft Corporation, uma corporação multinacional de tecnologia americana sediada em Redmond, Washington2.
* **GOOGL**: ativo da Google LLC, uma empresa multinacional de tecnologia americana.



```python
# Definindo os ativos e o período de tempo
ativos = ['AAPL', 'MSFT', 'GOOGL']
data_inicio = (datetime.now() - timedelta(days=2*365)).strftime('%Y-%m-%d')  # 2 anos atrás
data_fim = datetime.now().strftime('%Y-%m-%d')  # data atual
```

### 1.2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
# Baixando os dados
dados_ativos = {ativo: yf.download(ativo, start=data_inicio, end=data_fim) for ativo in ativos}

# Salvando os dados em um único DataFrame
df = pd.concat(dados_ativos, names=['Ativo', 'Data']).reset_index()
df.to_csv('ativos.csv', index=False)
```

## 2. Observando a volatilidade dentro do dia


### 2.1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).


```python
# Lendo os dados do arquivo CSV
df = pd.read_csv('ativos.csv')
```

```python
# Calculando a amplitude de variação dentro do dia para cada ativo
df['Amplitude'] = df['High'] - df['Low']
df
```

### 2.2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.


```python
# Criando o gráfico com Plotly
fig = go.Figure()
for ativo in df['Ativo'].unique():
    fig.add_trace(go.Scatter(x=pd.to_datetime(df[df['Ativo'] == ativo]['Data']),
                             y=df[df['Ativo'] == ativo]['Amplitude'],
                             mode='lines',
                             name=ativo))

# Ajustando a formatação do eixo x para mostrar todos os meses
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")

fig.show()
```

### 2.3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Calculando a média da amplitude para cada ativo
media_amplitude = df.groupby('Ativo')['Amplitude'].mean()

# Identificando o ativo com a maior média de amplitude
ativo_maior_volatilidade = media_amplitude.idxmax()

print(f'O ativo com a maior volatilidade é: {ativo_maior_volatilidade}')
```

Análise:<br>
 > Além de certificar com esse código, que calcula a média da amplitude para cada ativo e identifica o ativo com a maior média de amplitude (que é o ativo com a maior volatilidade), é possível verifificar claramente pela linha vermelha do ativo MSFT, da Microsoft, que é o mais volátil entre os três verificados.
