#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.

##### 2.1. DataFrame para o metadados:

````python
metadata = pd.DataFrame({'variavel': df.columns, 'tipo': df.dtypes})
metadata
````

##### 2.2. Adicionar uma coluna nesse *dataframe* chamada "qtd_categorias" e colocar o número de categorias correspondente de cada variável:

````python
# Calcular a quantidade de categorias para cada variável
qtd_categorias = []
for var in df.columns:
    qtd_categorias.append(df[var].nunique())

# Adicionar a quantidade de categorias ao metadados
metadata['qtd_categorias'] = qtd_categorias
````

````python
metadata
````

##### 2.3. Criar variáveis dummy para as variáveis necessárias:

````python
# Criar variáveis dummy para as variáveis categóricas
for var in df.columns:
    if df[var].dtype == 'object':
        dummies = pd.get_dummies(df[var], prefix=var)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(var, axis=1)
df.columns
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Criar um novo DataFrame para armazenar as variáveis numéricas e dummy
df_novo = pd.DataFrame()

# Para cada variável no DataFrame original:
for var in df.columns:
    # Se a variável é numérica ou booleana, copie-a para o novo DataFrame:
    if df[var].dtype in ['int64', 'float64', 'bool']:
        df_novo[var] = df[var]
    # Se a variável é categórica;
    elif df[var].dtype == 'object':
        # Cria variáveis dummy para essa variável:
        dummies = pd.get_dummies(df[var], prefix=var)
        # Adiciona as variáveis dummy ao novo DataFrame:
        df_novo = pd.concat([df_novo, dummies], axis=1)

df_novo
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# Criar tabelas de contingência:
contingencia_email = pd.crosstab(df['mau'], df['possui_email'])
contingencia_veiculo = pd.crosstab(df['mau'], df['posse_de_veiculo'])
````

````python
print("Tabela de contingência para 'possui_email':")
contingencia_email
````

````python
print("Tabela de contingência para 'posse_de_veiculo':")
contingencia_veiculo
````

Cada tabela mostra a distribuição da variável resposta mau para cada categoria das variáveis.

Para determinar qual variável é mais poderosa, é necessário analisar a  distribuição da variável resposta em cada categoria das variáveis preditoras. A variável que mostra a maior diferença na distribuição da variável resposta entre suas categorias pode ser considerada mais poderosa.



> Para a variável possui_email:

Dos clientes que não possuem e-mail (0), 360 são mau = True e 14.810 são mau = False.

Dos clientes que possuem e-mail (1), 30 são mau = True e 1.450 são mau = False.





> Para a variável posse_de_veiculo:

Dos clientes que não possuem veículo (N), 247 são mau = True e 9.931 são mau = False.

Dos clientes que possuem veículo (Y), 143 são mau = True e 6.329 são mau = False.



###### Agora, calculando a taxa de mau = True para cada categoria:


````python
# Taxa de mau = True para clientes sem e-mail:
360 / (360 + 14810)
````

````python
# Taxa de mau = True para clientes com e-mail:
30 / (30 + 1450)
````

````python
#Taxa de mau = True para clientes sem veículo:
247 / (247 + 9931)
````

````python
# Taxa de mau = True para clientes com veículo:
143 / (143 + 6329)
````



Observando essas taxas, nota-se que a variável possui_email mostra uma diferença maior na taxa de mau = True entre suas categorias em comparação com a variável posse_de_veiculo.
Portanto, com base nesses dados, pode-se dizer que a variável possui_email parece ser mais poderosa para prever a probabilidade de mau = True.
