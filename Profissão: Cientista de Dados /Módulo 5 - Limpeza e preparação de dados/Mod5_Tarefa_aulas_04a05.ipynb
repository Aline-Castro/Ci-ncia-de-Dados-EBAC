#### Marcação de bom e mau
O objetivo da modelagem é classificar o risco de inadimplência, ou como se diz no meio, o risco de *default*. Podemos fazer longas discussões sobre o conceito de *default* com base em estudos e exigências regulatórias, para efeitos deste estudo, um cliente em *default* é aquele que está em 60 dias de atraso ou mais. Então classificaremos os clientes como 'bons' e 'maus' assim:
- **Maus** pagadores: são aqueles que entraram em 'default' (atraso 60 dias ou mais) nos 24 meses seguintes à aquisição do cartão de crédito.
- **Bons** pagadores: são considerados todos os demais.
- **Excluídos**: Clientes que não adquiriram um cartão de crédito (seja por recusa, seja por desistência) não possuem informações de pagamento, portanto não se pode identificar se são bons ou maus.  ## 1) Tarefa: Marcar default no mês
### Faça uma indicadora de se o cliente está em default em cada uma das marcações (mes_00 a mes_24).
### Dica: você pode utilizar o método .isin() do Pandas.

> ###### A tarefa envolve a criação de uma variável indicadora para cada mês (de mes_0 a mes_24) para marcar se um cliente está em default ou não. Isso significa que, para cada mês, foi criado uma nova coluna no DataFrame pg que indique se o cliente está em default (atraso de 60 dias ou mais) ou não.



> ###### Para fazer isso, foi usado o método .isin() do Pandas para verificar se o status do mês está na lista de status que indicam default. O resultado é uma série booleana (True/False),convertida em inteiros (1/0) usando .astype(int) para criar a coluna indicadora.

````python
propostas = pd.DataFrame(propostas)
pg = pd.DataFrame(pg)

# Lista de status que indicam default
default_status = ['2', '3', '4', '5']

# Loop através de cada mês
for i in range(25):
    month = 'mes_{}'.format(i)
    # Cria uma nova coluna indicadora para cada mês
    pg[month + '_default'] = pg[month].isin(default_status).astype(int)
````

## Tarefa 2) 'bons' e 'maus' ao longo de todos os 24 meses de desempenho
### Marque para cada cliente se ele teve pelo menos um episódio de default entre o mês 0 e o mês 24.
### Dica: o método sum() pode ajudar. Caso precise, consulte a documentação e procure pelo argumento axis, você viu outros métodos que possuem esse argumento também. Tendo o número de meses em default de cada cliente, basta marcar True para todos aqueles que possuem pelo menos 1 mês em default e False para os demais.

> ###### A Tarefa 2 pede para marcar cada cliente como ‘bom’ ou ‘mau’ com base em se ele teve pelo menos um episódio de default entre o mês 0 e o mês 24. Assim, isso significa criar uma nova coluna no DataFrame pg que indique se o cliente teve pelo menos um episódio de default durante os 24 meses de desempenho.

> ##### Para fazer isso, foi usado o método .sum() do Pandas para calcular o número de meses em default para cada cliente. O resultado é uma nova coluna 'num_default_months' que contém o número de meses em default para cada cliente. Em seguida, foi criado uma nova coluna indicadora 'is_default' que é igual a True se o cliente teve pelo menos um episódio de default (ou seja, se 'num_default_months' é maior que 0) e False caso contrário.



````python

# Lista de colunas indicadoras de default
default_cols = ['mes_{}_default'.format(i) for i in range(25)]

# Número de meses em default para cada cliente
pg['num_default_months'] = pg[default_cols].sum(axis=1)

# Nova coluna indicadora para marcar se o cliente teve pelo menos um episódio de default
pg['is_default'] = pg['num_default_months'] > 0
````

## Tarefa 3) Marcando proponentes expostos ao risco de crédito
###Marcando proponentes que se tornaram tomadores: lembre-se de que clientes que não adquiriram o cartão devem ser desconsiderados. A base de pagamentos possui apenas clientes que adquiriram cartão de crédito, então você pode selecionar somente os clientes da base de propostas que se encontram na base de pagamentos.



> ###### Para marcar os proponentes que se tornaram tomadores de crédito, foi selecionado apenas os clientes da base de propostas que também estão presentes na base de pagamentos.


> ###### Isso foi feito usando o método .isin() do Pandas para verificar se o ID do cliente na base de propostas está presente na base de pagamentos. O resultado é um novo DataFrame propostas_tomadores que contém apenas os clientes da base de propostas que também estão presentes na base de pagamentos

````python
# 3) Seu código aqui
# Seleciona apenas os clientes da base de propostas que também estão presentes na base de pagamentos
propostas_tomadores = propostas[propostas['ID'].isin(pg['ID'])]
````
